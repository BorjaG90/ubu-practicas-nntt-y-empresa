#### Importación de librerías

```python
import networkx as nx # librería de análisis de redes
import numpy as np # librería de computación numérica
import matplotlib.pyplot as plt
#import plotly.plotly as py # Libreria para histogramas
%matplotlib inline
```

```python
def mostrar_pagerank(grafo):
    pos=nx.fruchterman_reingold_layout(grafo,k=.25)
    pk=np.array(nx.pagerank(grafo, alpha=0.85).values())
    print "PageRank"
    nx.draw(grafo,pos,with_labels=True,node_color='g', node_size=((pk*100)+5)**3, alpha=0.8,arrows=False)
```

```python
def mostrar_hits_hubs(grafo):
    pos=nx.nx.fruchterman_reingold_layout(grafo,k=.25)
    [H,A]=nx.hits(grafo)
    H=np.array(H.values())
    print "Hubs"
    nx.draw(grafo,pos,with_labels=True,node_color='b', node_size=((H*100)+5)**3, alpha=0.8,arrows=False)
```

```python
def mostrar_hits_auth(grafo):
    pos=nx.fruchterman_reingold_layout(grafo,k=.8)
    [H,A]=nx.hits(grafo)
    A=np.array(A.values())
    print "Authority"
    nx.draw(grafo,pos,with_labels=True,node_color='r', node_size=((A*100)+5)**3, alpha=0.8,arrows=False)
```

```python
G=nx.read_pajek("football.net")
G=nx.DiGraph(G)
```

## 1 Red de football

```python
mostrar_pagerank(G)
```

```python
mostrar_hits_hubs(G)
```

```python
mostrar_hits_auth(G)
```

## 2 Red de familias de florentinas del renacimiento

```python
#Eigenvector_centrality

G2=nx.florentine_families_graph()

Evect_networkx=np.array(nx.centrality.eigenvector_centrality(G2).values())
Evect_networkx_med=Evect_networkx/Evect_networkx.max()

nx.draw(G2,with_labels=True,node_color='r',alpha=0.3,node_size=Evect_networkx*5000)
```

```python
#Eigenvector_centrality (med)
nx.draw(G2,with_labels=True,node_color='b',alpha=0.3,node_size=Evect_networkx_med*3000)
```

```python
# Closeness centrality
closeness=np.array(nx.centrality.closeness_centrality(G2).values())
closeness_med=closeness/closeness.max()
# Visualización: tamaño de los nodos (closeness centrality)
nx.draw(G2,with_labels=True,node_color='g',alpha=0.3,node_size=closeness*5000)
```

```python
# Visualización: tamaño de los nodos (closeness centrality med)
nx.draw(G2,with_labels=True,node_color='g',alpha=0.3,node_size=closeness_med*3000)
```

```python
# Betweenness centrality
betweenness=np.array(nx.centrality.betweenness_centrality(G2).values())
# Número de caminos cortos (betweenness sin normalizar)
betweenness*((G2.number_of_nodes()-1)*(G2.number_of_nodes()-2))/2
# Visualización: tamaño de los nodos (betweenness centrality)
nx.draw(G2,with_labels=True,node_color='y',alpha=0.3,node_size=betweenness*5000)
```

```python
# Katz centrality
katz=np.array(nx.centrality.katz_centrality(G2).values())
nx.draw(G2,with_labels=True,node_color='b',alpha=0.3,node_size=katz*5000)
```
